### Exercise 1:

From the datasets folder, load the "tamiami.csv" file as a dataframe. Rename the columns (in order) to the following:

- location
- sales
- employees
- restaurants
- foodcarts
- price

Then do a correlation table on that dataframe. What features (columns) are correlated? What features aren't correlated?

```python
import pandas as pd
import numpy as np
```

```python
location = 'datasets/tamiami.csv'
df = pd.read_csv(location)
```

```python
df.columns
```

```python
df.rename(columns={'Cart Location' : 'location', 'Hot Dog Sales' : 'sales', 
                   'Employees in Nearby Office Buildings' : 'employees', 
                   'Num of Nearby Restaurants' : 'restaurants', 
                   'Num of Other Food Carts Nearby' : 'foodcarts', 
                   'Price' : ' price'}, inplace=True)
df.head()
```

```python
df.corr()
#Features Correlated: 
    #(location & Sales), (location & restaurants), (locations & foodcarts)
    #(sales & employees), (sales & restaurants)
    #(employees & restaurants)
    #(foodcart & price)
#Features Not Correlated: 
    #(location & employees), (location & price)
    #(sales & foodcarts), (sales & price)
    #(employees & foodcarts), (employees & price)
    #(restaurants & foodcarts), (restaurants & price)
```

### Exercise 2:

Using the dataframe from the previous exercise, choose features (columns) to create a linear regression formula to predict sales. Try it with and without the y-intercept. How does it make a difference? Does adding or removing features in your model formula make a difference in the output?

```python
import statsmodels.formula.api as smf
```

```python
df.columns
```

```python
result = smf.ols('sales ~ location + employees + restaurants + foodcarts', data=df).fit()
```

```python
result.summary()
```

```python
result = smf.ols('sales ~ location + employees + restaurants + foodcarts - 1', data=df).fit()
```

```python
result.summary()
```

```python
result = smf.ols('sales ~ employees + restaurants + foodcarts', data=df).fit()
```

```python
result.summary()
```

```python
result = smf.ols('sales ~ location + employees + restaurants', data=df).fit()
```

```python
result.summary()
```
